```python
import numpy as np
import folium
from folium.plugins import HeatMap
import matplotlib.pyplot as plt
import pandas as pd

m = folium.Map([31.262,121.4808], zoom_start=10)
```

```python
def plot_circle_on_map(df_row, m):
    folium.Marker(
        location= [float(df_row['CordLongitude']), float(df_row['CordLatitude'])],
        popup=df_row['Name'],
        icon=folium.Icon(color="lightred", icon="info-sign"),
    ).add_to(m)

    folium.CircleMarker(
        location= [float(df_row['CordLongitude']), float(df_row['CordLatitude'])],
        radius = 10,
        color = '#ff80ff',
        fill=False
    ).add_to(m)
```

```python
df6 = pd.read_csv('../data6.csv')
df6.head(5)

for i, df_row in df6.iterrows():
    plot_circle_on_map(df_row, m)

m
```
