```python
import numpy as np
import folium
from folium.plugins import HeatMap
import matplotlib.pyplot as plt
import pandas as pd

m = folium.Map([31.262,121.4808], zoom_start=14)
```

```python
def popup_html(df_row):
    html = """
    <!DOCTYPE html>
    <html>
    <center><h4 style="margin-bottom:5"; width="200px">{}</h4>""".format(df_row['Name']) + """</center>
    <center><p>积分{}""".format(df_row['RawScore']) + """</p></center>    
    <center><p>50分需要社保月数: {}""".format(int(df_row['Month50Base'])) + """</p></center>
    </html>
    """

    return html
```

```python
def calc_color(df_row):
    t = (int(df_row['Month50Base']))
    rgb = (255-t,255-t,255-t)
    return '#%02x%02x%02x' % rgb
```

```python
def plot_circle_on_map(df_row, m):
    html = popup_html(df_row)
    popup = folium.Popup(folium.Html(html, script=True), max_width=500)

    folium.Marker(
        location= [float(df_row['CordLongitude']), float(df_row['CordLatitude'])],
        popup=popup,
        icon=folium.Icon(color="lightred", icon="info-sign"),
    ).add_to(m)

    folium.CircleMarker(
        location= [float(df_row['CordLongitude']), float(df_row['CordLatitude'])],
        radius = 2 + (int(df_row['Month50Base'])) / 20,
        color = calc_color(df_row),
        fill=True
    ).add_to(m)
```

```python
df6 = pd.read_csv('../data6.csv')

for i, df_row in df6.iterrows():
    plot_circle_on_map(df_row, m)

m
```
